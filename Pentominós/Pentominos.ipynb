```python
#Implementamos como una instancia de la clase ProblemaEspacioEstados un problema de espacio de estados, 
#proporcionando una lista de acciones, un estado inicial y un método para comprobar estados finales.
class Pentominos(probee.ProblemaEspacioEstados):
    def __init__(self, filas, cols, nMaxI):
        #Nos creamos la matriz
        matriz = np.zeros((filas,cols), dtype=np.str_)
        
        #Nos creamos una tupla con los nombres de las acciones (tupla de String)
        acciones=[]
        cad='poner_Figura_'
        ac=[]
        letras = ['I','IH','L','Z','N','P','T','U','V','W','X','Y','F']#
        for a in letras:
            res = cad+a
            ac.append(res)
        
        #Llamamos a las acciones creadas en fichas.py mediante la tupla de String creada anteriormente
        for n in ac:
            for j in range(0,cols):
                for i in range(0,filas):
                    cad2 = 'fichaas.'+n+'(i,j)'
                    #El metodo eval() evalua un String
                    acciones = acciones + [eval(cad2)]  
        
        estado_inicial = [matriz, filas*cols, nMaxI] 
        super().__init__(acciones, estado_inicial)        
    
    def es_estado_final(self, estado): 
        return estado[1] <= 4 #Tiene que haber 4 espacios
```
